# training data 만들기

```python
img_size=50
data=[]
path='./Medical Mask/images/'     
```

```python
train
```

## 이미지 전처리(crop+resize)

```python
def create_data():
       for i in range(len(train)):
            arr = []
            for j in train.iloc[i]:
                   arr.append(j)
            img_array = cv2.imread(os.path.join(images, arr[0]), cv2.IMREAD_GRAYSCALE)
            if img_array is None: #train data는 없는 이미지가 있음
                continue
            crop_image = img_array[arr[2]:arr[4], arr[1]:arr[3]]
            new_img_array = cv2.resize(crop_image,(img_size,img_size))
            data.append([new_img_array, arr[5]]) #전처리 이미지와 분류클래스
            #print(f"{i}까지 완료")
create_data() 
```

```python
data[0][0]
plt.imshow(data[0][0])
```

```python
x = []
y = []
cnt = 0
for features, labels in data:
    if cnt == 0: #첫번째는 face_no_mask
        print(features)
        print(labels)
    x.append(features)
    y.append(labels)
    cnt += 1
```

```python
#범주형 데이터를 숫자형 데이터로 인코딩
from sklearn.preprocessing import LabelEncoder

lbl = LabelEncoder()
y = lbl.fit_transform(y)
```

```python
y #face_no_mask는 0, face_with_mask는 1
```

```python
x = np.array(x).reshape(-1,50,50,1)
x = tf.keras.utils.normalize(x, axis=1)

# from keras.utils import to_categorical
from tensorflow.keras.utils import to_categorical

y = to_categorical(y)
```

```python
y
```

# Model 학습

```python
model=Sequential()
model.add(Conv2D(100,(3,3), input_shape = x.shape[1:], activation = 'relu', strides=2))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(50, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(2, activation = 'softmax'))
```

```python
class_weight = {0: 1.0, 1: 2.0}
opt = tf.keras.optimizers.Adam(lr=1e-3, decay=1e-5)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy']) 
model.fit(x, y, epochs=30, batch_size=5, class_weight = class_weight)
```

test image의 bounding-box를 만들기 위해 얼굴을 찾는 MTCNN을 사용

```python
detector = MTCNN()
img = plt.imread(os.path.join(images,test_images[0]))
face = detector.detect_faces(img)
print(face)
for face in face:
        bounding_box = face['box']
        x = cv2.rectangle(img, (bounding_box[0], bounding_box[1]),
              (bounding_box[0]+bounding_box[2], bounding_box[1] + bounding_box[3]), (0,155,255), 10)
        plt.imshow(x)
```

```python
img = plt.imread(os.path.join(images,test_images[3]))
face = detector.detect_faces(img)
for face in face:
        bounding_box = face['box']
        x = cv2.rectangle(img, (bounding_box[0], bounding_box[1]),
              (bounding_box[0] + bounding_box[2], bounding_box[1] + bounding_box[3]), (0,155,255), 10)
        plt.imshow(x)
```

```python
detector = MTCNN()
test_df = []
for image in test_images:
    img = plt.imread(os.path.join(images,image))
    faces = detector.detect_faces(img)
    #[{'box': [1621, 241, 2772, 3446], 'confidence': 0.9986934065818787, 'keypoints': {'left_eye': (2695, 1625), 'right_eye': (3817, 1570), 'nose': (3393, 2347), 'mouth_left': (2706, 2891), 'mouth_right': (3529, 2929)}}]
    test = []
    for face in faces:
        bounding_box = face['box']
        test.append([image, bounding_box]) #이미지데이터와 박스 좌표를 test에 넣는다.
    test_df.append(test) #test 리스트를 test_df에 넣는다.
```

```python
#새로 만든 test에 이미지를 담는다.
test = []
for i in test_df:
    if len(i) > 0:
        if len(i) == 1:
            test.append(i[0])
        else:
            for j in i:
                test.append(j)
```

```python
sub = []
rest_image = []
for i in test:
    sub.append(i[0])
for image in test_images:
    if image not in sub:
        rest_image.append(image)      
```

```python
detector = MTCNN()
test_df_ = []
for image in rest_image:
    img = cv2.imread(os.path.join(images,image))
    faces = detector.detect_faces(img)
    test_ = []
    for face in faces:
        bounding_box = face['box']
        test_.append([image,bounding_box])
    test_df_.append(test_) 
for i in test_df_:
    if len(i) > 0:
        if len(i) == 1:
            test.append(i[0])
        else:
            for j in i:
                test.append(j) 
```

```python
test_data=[]
def create_test_data():
            for j in test:
                if j not in negative:
                    img = cv2.imread(os.path.join(images,j[0]), cv2.IMREAD_GRAYSCALE)
                    img = img[j[1][1]:j[1][1]+j[1][3], j[1][0]:j[1][0]+j[1][2]]
                    new_img = cv2.resize(img,(50,50))
                    new_img = new_img.reshape(-1,50,50,1)
                    predict = model.predict(new_img)
                    print(predict)
                    test_data.append([j,predict])

create_test_data()      
```

```python
image = []
classname = []
for i,j in test_data:
    classname.append(np.argmax(j))
    image.append(i)
df = pd.DataFrame(columns = ['image','classname'])
df['image'] = image
df['classname'] = classname
df['classname'] = lbl.inverse_transform(df['classname'])
df
```
